### 1. use of create task and geather (Basic Format)

````python
import time
import asyncio

# Define a coroutine
async def greet(name):
    print(f"Hello, {name}!")
    await asyncio.sleep(1)  # Simulate an I/O-bound operation
    print(f"Goodbye, {name}!")

# Define the main coroutine
async def main():
    create_task =True
    use_gather = False

    if create_task:
        # Create tasks for concurrent execution
        task1 = asyncio.create_task(greet("Alice"))
        task2 = asyncio.create_task(greet("Bob"))

        # Wait until both tasks are completed
        await task1
        await task2

    elif use_gather:
        # use of gather
        tasks = [greet("Alice"),greet("Bob")]
        await asyncio.gather(*tasks)
        #await asyncio.gather(asy_sleep(2), asy_sleep(1), print_hello())






if __name__ == "__main__":
    VSCODE = False
    JUPYTER = True
 
    if VSCODE:
        # Run the event loop
        asyncio.run(main())
        
    elif JUPYTER:
        s = time.perf_counter()
        await main()
        elapsed = time.perf_counter() - s
        print(f"Script executed in {elapsed:0.2f} seconds.")
````

##### both the execution time is equal why
    Key Differences in Execution
    Blocking vs. Non-Blocking Operations:

    Sequential Code: Uses time.sleep(2), which is a blocking call. The entire program waits for 2 seconds before proceeding to the next line of code.
    Asyncio Code: Uses await asyncio.sleep(2), which is a non-blocking call within an asynchronous event loop. However, if the tasks are awaited sequentially, the total wait time accumulates just like in the blocking case.
    Concurrency vs. Parallelism:

    Sequential Code: Executes each task one after the other, blocking on each time.sleep(2) call.
    Asyncio Code: Can run tasks concurrently, but if the tasks are awaited one by one (in a sequential manner), it effectively behaves like sequential execution.

    Concurrency vs. Parallelism: If you use await asyncio.sleep(2) in a non-blocking manner but only run one task at a time, the program waits for each task to complete before moving to the next. This is similar to the sequential execution case.

    True Concurrency: To take full advantage of asyncio, you need to run multiple tasks concurrently without waiting for each one to complete before starting the next. This is achieved by using asyncio.gather to execute all tasks at once.

````python
import asyncio
import time

async def fn(n):
    await asyncio.sleep(2)  # Corrected: Added await
    return n*n

async def main(N):
    results = []  # List to store all results
    tasks = []
    for i in range(N):
        num = (i + 1) * 10000
        tasks.append(fn(num))  # Create a task
        

    results = await asyncio.gather(*tasks)
    
    for result in results:
        print(result)
   
        
    return results  # Return the list of all results

VSCODE = False
JUPYTER = True

if VSCODE:
    if __name__ == "__main__":
        asyncio.run(main(4))
        
elif JUPYTER:
    s = time.perf_counter()
    results = await main(4)  # Await the main function and get the results
    elapsed = time.perf_counter() - s
    print(f"Script executed in {elapsed:0.2f} seconds.")
````

### 5. handle producer and consumer

````python
import asyncio
import time

async def producer(queue):
    for i in range(5):
        await asyncio.sleep(1)
        await queue.put(f"Item {i}")
        print(f"Produced Item {i}")

async def consumer(queue):
    while True:
        item = await queue.get()
        if item is None:  # Poison pill to stop the consumer
            break
        print(f"Consumed {item}")
        queue.task_done()

async def main():
    queue = asyncio.Queue()
    
    producer_task = asyncio.create_task(producer(queue))
    consumer_task = asyncio.create_task(consumer(queue))
    
    await producer_task
    await queue.put(None)  # Poison pill to stop the consumer
    await consumer_task

VSCODE = False
JUPYTER = True

if VSCODE:
    if __name__ == "__main__":
        asyncio.run(main())
        
elif JUPYTER:
    s = time.perf_counter()
    await main()  # Await the main function and get the results
    elapsed = time.perf_counter() - s
    print('\n')
    print(f"Script executed in {elapsed:0.2f} seconds.")

````

### 7. even though there are 5 tasks, only 2 will run concurrently due to the semaphore 

````python
#In this example, even though there are 5 tasks, only 2 will run concurrently due to the semaphore.


import asyncio

sem = asyncio.Semaphore(2)  # Limit concurrent tasks to 2

async def limited_task(n):
    async with sem:
        print(f"Task {n} started")
        await asyncio.sleep(2)
        print(f"Task {n} finished")

async def main():
    tasks = [limited_task(i) for i in range(5)]
    await asyncio.gather(*tasks)

VSCODE = False
JUPYTER = True

if VSCODE:
    if __name__ == "__main__":
        asyncio.run(main())
        
elif JUPYTER:
    s = time.perf_counter()
    await main()  # Await the main function and get the results
    elapsed = time.perf_counter() - s
    print('\n')
    print(f"Script executed in {elapsed:0.2f} seconds.")
````

# TCP server vs HTTP server

A **TCP server** is a type of network server that communicates with clients using the **Transmission Control Protocol (TCP)**. TCP is one of the core protocols of the internet protocol suite and provides reliable, ordered, and error-checked delivery of data between applications running on hosts over a network.

### Key Features of a TCP Server:

1. **Connection-Oriented**: TCP is connection-oriented, meaning that before the server and client can communicate, they must establish a connection (handshake) first. This ensures a reliable and ordered stream of data.

2. **Reliable Communication**: TCP guarantees that data will be delivered accurately and in the correct order. If any packets are lost during transmission, TCP will automatically request retransmission until the data arrives correctly.

3. **Stream-based**: Data is transmitted as a continuous stream of bytes. TCP handles breaking this stream into packets, transmitting them, and reassembling them at the destination.

4. **Full-Duplex Communication**: TCP allows for bidirectional communication, meaning both the server and client can send and receive data at the same time.

### How a TCP Server Works:

1. **Listening for Connections**: 
   - The server listens on a specific IP address and port for incoming connection requests from clients. This is done using a **socket**, which represents the communication endpoint.
   
2. **Establishing a Connection**: 
   - When a client attempts to connect, the server accepts the connection, establishing a dedicated communication channel between the server and the client.
   
3. **Exchanging Data**: 
   - Once connected, the server and client can exchange data in both directions. The server can receive data from the client, process it, and send a response back.

4. **Terminating the Connection**: 
   - After the communication is complete, either the server or the client can close the connection, terminating the session.

### Key Steps in a TCP Server:

1. **Bind to a Port**: The server binds to a specific IP address and port so that clients know where to connect.
2. **Listen for Incoming Connections**: The server enters a "listening" state, waiting for connection requests from clients.
3. **Accept Connections**: Once a client initiates a connection, the server accepts it and can start communicating.
4. **Send and Receive Data**: The server and client exchange data as needed. The server processes client requests and sends back responses.
5. **Close the Connection**: The connection is closed once the communication is finished.

---

### Example of How a TCP Server Works:

#### Server-Side Code:
```python
import socket

# Create a TCP/IP socket
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# Bind the socket to an address and port
server_address = ('127.0.0.1', 65432)  # Localhost and port 65432
server_socket.bind(server_address)

# Listen for incoming connections
server_socket.listen(5)  # Can accept up to 5 connections

print(f"Server listening on {server_address}")

while True:
    # Wait for a connection
    connection, client_address = server_socket.accept()
    
    try:
        print(f"Connection from {client_address}")
        
        # Receive data in small chunks and send it back
        while True:
            data = connection.recv(1024)
            if data:
                print(f"Received: {data.decode()}")
                connection.sendall(data)  # Echo the data back to the client
            else:
                break
    
    finally:
        # Clean up the connection
        connection.close()
```

#### Client-Side Code:
```python
import socket

# Create a TCP/IP socket
client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# Connect to the server
server_address = ('127.0.0.1', 65432)
client_socket.connect(server_address)

try:
    # Send some data
    message = 'Hello, TCP Server!'
    client_socket.sendall(message.encode())
    
    # Wait for the response
    data = client_socket.recv(1024)
    print(f"Received: {data.decode()}")
    
finally:
    # Close the connection
    client_socket.close()
```

### Explanation:
- **Server**: The server listens on `127.0.0.1` at port `65432`. When a client connects, the server accepts the connection and waits for the client to send data. The server reads the data, processes it (in this case, echoes it back), and then continues to listen for new connections.
  
- **Client**: The client connects to the server using the same IP address and port. After establishing the connection, it sends a message, receives a response, and then closes the connection.

### Real-World Examples of TCP Servers:
- **Web Servers**: HTTP and HTTPS protocols run on TCP. When you visit a website, your browser (the client) establishes a TCP connection to the web server.
- **FTP Servers**: File Transfer Protocol uses TCP to ensure reliable file transfers between the client and the server.
- **Email Servers**: SMTP (Simple Mail Transfer Protocol) and IMAP/POP (for email retrieval) use TCP to ensure reliable delivery of email data.

---

### Summary:
A **TCP server** is responsible for accepting incoming client connections, processing the data sent by the clients, and sending back responses, all while ensuring reliable communication through the use of the TCP protocol. The server can handle multiple clients simultaneously, ensuring that data is delivered in the correct order and without errors.

## TCP server Flask server

Yes, a **TCP server** is different from a server created using a **Flask API**, though both facilitate communication between clients and servers. The key difference lies in their **protocols, purpose**, and **level of abstraction**. Let's break this down:

### 1. **TCP Server**:
- **Protocol**: Uses the **Transmission Control Protocol (TCP)**, which operates at a low level in the networking stack.
- **Purpose**: A TCP server is designed to handle low-level communication between two systems over a network. It works with **raw byte streams** and requires developers to handle the connection, transmission, and termination manually.
- **Level of Abstraction**: TCP servers provide more control over how data is sent and received, but they require you to handle things like encoding/decoding data, managing connection lifecycles, and ensuring data reliability.

### 2. **Flask Server (HTTP Server)**:
- **Protocol**: Flask operates on top of the **HTTP/HTTPS protocols**, which are application-level protocols built on top of TCP. HTTP is stateless, meaning each request from a client (e.g., a browser) is treated independently.
- **Purpose**: A Flask server is designed for building **web applications** and **REST APIs**. It provides a higher level of abstraction, allowing developers to define web routes (URLs), handle GET/POST requests, and manage client-server interactions easily.
- **Level of Abstraction**: Flask abstracts away the low-level networking operations, allowing developers to focus on defining the logic for routes, parsing data (e.g., JSON), and sending structured HTTP responses.

---

### Key Differences Between a TCP Server and a Flask API Server:

| Aspect             | TCP Server                                | Flask Server (HTTP)                         |
|--------------------|-------------------------------------------|---------------------------------------------|
| **Protocol**       | TCP (low-level)                           | HTTP/HTTPS (built on top of TCP)            |
| **Purpose**        | General communication, any kind of data exchange (e.g., file transfer, chat app). | Building web applications and RESTful APIs |
| **Level of Control**| Full control over the connection and data flow, but you handle everything manually (data parsing, encoding, etc.). | Abstracts low-level details, handles requests/responses based on URLs, and sends structured HTTP responses. |
| **Data Format**    | Typically byte streams, with encoding/decoding required. | Mostly deals with JSON, HTML, or other high-level formats. |
| **Connection Type**| Persistent connection, data can be sent in streams. | Stateless, each HTTP request is independent (though you can use sessions/cookies). |
| **Use Case**       | Low-level applications: chat servers, raw data transmission, custom protocols. | High-level web services: APIs, web applications, serving HTML, JSON, etc. |
| **Client**         | Any custom program that speaks TCP.       | Typically web browsers, or clients making HTTP requests (like `curl`, or `requests` in Python). |
| **Libraries**      | Requires handling socket programming.     | Uses the Flask framework, which abstracts network details and provides routing and request handling. |

---

# TCP server and scoket

Yes, a **TCP server** is different from **sockets**, though they are closely related. A **TCP server** uses **sockets** to facilitate communication over a network. To clarify the difference:

### 1. **Socket**:
A **socket** is a low-level networking interface that provides the connection point for communication between two machines over a network. Sockets are the building blocks for network communication in many protocols, including TCP and UDP.

- **Definition**: A socket is an endpoint for sending or receiving data over a network connection.
- **Types**: Sockets can be used with different protocols, such as **TCP** (Transmission Control Protocol) or **UDP** (User Datagram Protocol). For a **TCP** connection, we use **stream sockets** (which use TCP), while **datagram sockets** use UDP.
- **Bidirectional Communication**: Sockets facilitate full-duplex communication, meaning that data can be sent and received simultaneously.
- **Socket Operations**:
  - **Create**: A socket is created using the `socket()` function.
  - **Bind**: A socket can be bound to a specific IP address and port.
  - **Listen**: A socket can listen for incoming connections (if it's a server).
  - **Connect**: A socket can connect to a remote socket (if it's a client).
  - **Send and Receive**: Data can be sent and received through the socket.
  - **Close**: A socket can be closed once communication is done.

### 2. **TCP Server**:
A **TCP server** is a higher-level application that **uses sockets** to establish a **TCP connection** and communicate with clients over a network.

- **Definition**: A TCP server listens for incoming TCP connections from clients, establishes a connection, and then communicates with the clients using sockets.
- **Protocol**: A TCP server specifically uses the **Transmission Control Protocol (TCP)**, which ensures reliable, ordered, and error-checked delivery of data.
- **Operation**: 
  - It listens on a specific IP address and port (using a socket).
  - When a client connects, it creates a new socket to communicate with that client.
  - It handles the communication (send/receive data) using this new socket.
  - When communication is complete, the socket is closed.

### Relationship Between TCP Server and Sockets:
- **Sockets are used by a TCP server** to manage network communication. Without sockets, a TCP server cannot send or receive data over the network.
- **TCP Server** is an application, while **sockets** are the underlying mechanism for network communication.

---

### Example: TCP Server Using Sockets

A TCP server **uses sockets** to manage network communication. Hereâ€™s an example that shows how a TCP server creates a socket and uses it to listen for client connections.

```python
import socket

# 1. Create a TCP/IP socket
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# 2. Bind the socket to an address and port
server_socket.bind(('127.0.0.1', 65432))

# 3. Listen for incoming connections
server_socket.listen(5)
print("Server listening on 127.0.0.1:65432")

while True:
    # 4. Accept a new connection
    connection, client_address = server_socket.accept()
    print(f"Connected to {client_address}")

    try:
        # 5. Receive data from the client
        data = connection.recv(1024)
        print(f"Received {data.decode()}")

        # 6. Send a response back to the client
        connection.sendall(b"Hello from server!")

    finally:
        # 7. Close the connection
        connection.close()
```

### Breakdown of Socket Usage in the TCP Server:
1. **Creating a Socket**: The server uses `socket()` to create a new socket (`socket.AF_INET` specifies the address family for IPv4, and `socket.SOCK_STREAM` specifies that the server will use the TCP protocol).
   
2. **Binding the Socket**: The socket is bound to an IP address and port using `bind()`. This tells the operating system where to listen for incoming connections.

3. **Listening for Connections**: The server calls `listen()` to put the socket in listening mode, where it waits for clients to connect.

4. **Accepting Connections**: When a client connects, the server calls `accept()`, which returns a new socket (`connection`) that is used to communicate with the client. The original socket (`server_socket`) continues to listen for new connections.

5. **Sending and Receiving Data**: The server uses `recv()` to receive data from the client and `sendall()` to send data back to the client.

6. **Closing the Connection**: Once the communication is complete, the server closes the connection using `close()`.

### In Summary:

- **Socket**: A low-level interface for communication between devices. It can be used for both TCP and other protocols (e.g., UDP). It represents the communication endpoint (either client or server).
- **TCP Server**: A server application that uses a socket to handle network communication over the TCP protocol. It listens for client connections, establishes a socket for each connection, and uses it to send and receive data.
  
Thus, while **sockets** are the mechanism for communication, the **TCP server** is the application that uses them to manage communication with multiple clients.

### **What is a Session?**
A **session** in networking or web development refers to a series of interactions or communications between a client and a server that take place over a period of time. The session helps maintain **state** (i.e., keeping track of previous interactions or data) across multiple interactions.

- **Session in Networking**: In the context of a network (such as TCP), a session is an active connection between two devices, where data can be exchanged back and forth.
- **Session in Web Development**: In web development (e.g., using Flask), a session refers to data or state that the server temporarily stores to keep track of a userâ€™s activity between different HTTP requests, which would otherwise be independent and stateless.

---

### **Connection Types:**
#### 1. **Persistent Connection During the Session** (TCP Server):
TCP connections maintain **persistent sessions**.

- **Persistent connection**: Once a TCP connection is established between a client and server, that connection remains open (persistent) for the duration of the session. Data can be sent and received between the two parties continuously, as long as the connection is maintained.
- **Session in TCP**: 
  - When a client connects to a TCP server, a session is initiated, and the connection is established.
  - During this session, the client and server can exchange data multiple times without needing to re-establish the connection.
  - The session ends when either the client or server closes the connection.
  
#### Example of a TCP Session:
- **Step 1**: A client connects to a server (e.g., through a socket).
- **Step 2**: The server and client can continuously exchange messages (bi-directional communication).
- **Step 3**: The session ends when one side explicitly closes the connection (i.e., the socket is closed).

Since the connection is persistent, you can think of it like a **phone call**: once the call is established, both parties can talk back and forth without needing to reconnect each time.

---

#### 2. **Stateless, Each Request Is Independent** (Flask HTTP Server):
In contrast, **HTTP** (the protocol Flask uses) is **stateless**, meaning each request is treated as a separate and independent interaction.

- **Stateless protocol**: In HTTP, every request from a client to the server is independent of any other request. After the server responds, the connection between the client and the server is closed. The server doesnâ€™t retain any information about the client between requests (unless sessions or cookies are implemented to simulate state).
- **No persistent session**: Each HTTP request/response cycle does not maintain a continuous connection between the client and server. Once a request is served, the server closes the connection. If the client needs to send more data, it initiates a new request, and the server treats it as a brand-new connection.
  
#### Example of HTTP Requests:
- **Step 1**: A client (e.g., web browser) sends a GET request to an HTTP server (e.g., Flask).
- **Step 2**: The server processes the request and sends a response.
- **Step 3**: The connection between the client and server is closed after the response.
- **Step 4**: If the client wants more data, it sends a new request, and the server treats it as an independent interaction.

This is similar to a **series of emails**: each email (HTTP request) is independent, and the server doesnâ€™t retain any state about previous emails unless you use mechanisms like sessions or cookies to track user data.

---

### **Key Differences Between Persistent and Stateless Connections:**

| **Aspect**               | **Persistent Connection (TCP)**      | **Stateless Connection (HTTP)**        |
|--------------------------|--------------------------------------|----------------------------------------|
| **Connection Type**       | Continuous, remains open during the session | Each request/response is independent   |
| **Session Management**    | Session lasts as long as the connection is open | Session information must be passed explicitly |
| **State Management**      | Connection retains state, no need to reconnect | Stateless, no state is kept between requests unless explicitly managed |
| **Use Cases**             | Real-time communication, streaming, chat apps | Web apps, REST APIs, websites         |

### Example:
- **TCP Session**: 
  - Client connects to a chat server, and they exchange messages. The connection remains open, and both client and server maintain state throughout the conversation. The client doesn't need to re-connect for each message.
  
- **HTTP Stateless**:
  - A client sends a request to a Flask server to fetch a webpage. The server processes the request and sends the HTML response. The connection is then closed. If the client clicks another link, a new request is made, and the server treats it independently from the previous request.

---

### **Simulating State in HTTP (Flask):**

Even though HTTP is stateless, **sessions** can be simulated using mechanisms such as **cookies** or **session objects** in Flask. These allow the server to keep track of a userâ€™s activity across multiple requests by storing session data.

#### Example of Flask Session:
```python
from flask import Flask, session

app = Flask(__name__)
app.secret_key = 'secret_key'

@app.route('/')
def index():
    # Increment a visit counter in the session
    if 'visits' in session:
        session['visits'] = session.get('visits') + 1
    else:
        session['visits'] = 1
    return f"Number of visits: {session['visits']}"

if __name__ == '__main__':
    app.run()
```

- **Session in Flask**: In this example, the server keeps track of how many times a user has visited the page by storing this information in the session object. Although HTTP is stateless, Flask uses a cookie to store the session ID, which links the user to their data stored on the server.

---

### **Summary**:
- **TCP (Persistent Connection)**: The connection between client and server is continuous, meaning data can flow back and forth as long as the session is open. It's ideal for scenarios that require ongoing communication, such as chat apps or video streaming.
  
- **HTTP (Stateless Connection)**: Each HTTP request is independent of others, meaning the server doesn't retain any memory of previous requests. This makes HTTP ideal for web apps and APIs. To maintain session-like behavior in HTTP, you must use mechanisms like cookies or session management techniques.
