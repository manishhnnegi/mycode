# Character Level Language modeling using MLP (Multi Layer Perceptron)

### Dataset:
    dataset a text file <names.txt> having around 32033 different names in english language.

### Archtecture of MLP:
    1. C embedding look up metrix
    2. 1st hidden layer having N no of neurons [input:3 output:N]
    3. tanh
    3. 2nd hidden layer having 27 no of neurons [input:N output:27] (27: vocab size)
    4. softmax 
    


### Goal:
    To build and train character level Language modeling using Neural network (MLP).
    which will generate the new names after learning.
    

### Summary Points:
  
    ## Total vocab size is 27
        26 unique charcters in english 
        One special chatacter "."

    ## Embedding Lookup table or metrix C:
    C is a metrix (vocab_size* N)
    for each character (part of vocab) we have unique integer 
    now we are mapping this integer to some embedding vector of dim N

    in our data:
    vocab_size = 27
    vector dim = 2
    C shape = {27,2}
    C will be having embedding vector of dim 2 for all the vocab from 0 to 26
    we will initalize it randomly 
    in training process its params will get updataed as happen in weights and baises.

    now let say transsformation of char a
    cha: a -> int: 1 -> vector: [0.4, 0.8]

### References:
    MLP paper:
    https://www.jmlr.org/papers/volume3/bengio03a/bengio03a.pdf

```python
# Build the vocabulary of characters and mappings to/from integers:
#--------------------------------------------
# in this dataset:
# Total vocab size is 27
    # 26 unique charcters in english 
    # One special chatacter "."




chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
print(f"indx to char:{itos}")
```

```python
# Build the dataset
#----------------------------
# given three charactyes model will predict the next character in the sequence:
# manish:
    # ... -> m
    # ..m -> a
    # .ma -> n
    # man -> i
    # ani -> s
    # nis -> h


block_size = 3 # context length: how many characters do we take to predict the next one?
X, Y = [], []
for w in words[:2]:

  print(w)
  context = [0] * block_size
  for ch in w + '.':
    ix = stoi[ch]
    X.append(context)
    Y.append(ix)
    print(''.join(itos[i] for i in context), '--->', itos[ix])
    context = context[1:] + [ix] # crop and append

    print(f"X: {X}")
    print(f"Y: {Y}")
  print(f"______________")
  print(f"______________")



X = torch.tensor(X)
Y = torch.tensor(Y)
```

```python
import matplotlib.pyplot as plt
# back to code
# optimisation for the firest 5 words in the text file



# Dataset creation:
# build the dataset with 5 words:
#----------------------------------------------------------------
block_size = 3 # context length: how many characters do we take to predict the next one?
X, Y = [], []
for w in words[:5]:
  #print(w)
  context = [0] * block_size
  for ch in w + '.':
    ix = stoi[ch]
    X.append(context)
    Y.append(ix)
    #print(''.join(itos[i] for i in context), '--->', itos[ix])
    context = context[1:] + [ix] # crop and append


X = torch.tensor(X)
Y = torch.tensor(Y)

print(f"Xshape:{X.shape}")
print(f"Yshape:{Y.shape}")
print(f"y type:{Y.dtype}")


# Params (wight & baises) initalization:
#----------------------------------------------------------------
g = torch.Generator().manual_seed(2147483647) # for reproducibility
C = torch.randn((27, 2), generator=g)
W1 = torch.randn((6, 100), generator=g)
b1 = torch.randn(100, generator=g)
W2 = torch.randn((100, 27), generator=g)
b2 = torch.randn(27, generator=g)
parameters = [C, W1, b1, W2, b2]


paramss = sum(p.nelement() for p in parameters) # number of parameters in total
#print(f"{paramss=}")


for p in parameters:
  p.requires_grad = True

# Training loop:
#-----------------------------------------------
import time
start = time.time()
epochs = 50
losses = []
for _ in range(epochs):

  #forward pass:
  emb = C[X] # (32, 3, 2)
  h = torch.tanh(emb.view(-1, 6) @ W1 + b1) # (32, 100)
  logits = h @ W2 + b2 # (32, 27)
  loss = F.cross_entropy(logits, Y)
  #print(f"loss: {loss}")

  # Store loss for this epoch
  losses.append(loss.item())
  # backwaed pass:
  for p in parameters:
    p.grad = None
  loss.backward()

  #wt updatae:
  for p in parameters:
    p.data += -0.1 * p.grad

end = round(time.time()-start, 2)

#---------------------------------
print(f"\n\nFinal results:\n")
print(f"-----------------------------------------")
print(f"total epochs: {epochs=}")
print(f"Total training time: {end} seconds")
print(f"final Loss: {loss.item()}")
print(f"Total Params:{paramss}")
print(f"-----------------------------------------")
# in batch of 32 the output results comparision 
# it has over fit for the middle examples
print(f"predicted labels and probs: \n {logits.max(1)=}")
print(f"True label: \n {Y=}")

# Plot loss vs epochs
plt.figure(figsize=(4, 3))  # Set the width and height in inches
plt.plot(losses)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```

```python
# Dataset:
block_size = 3 # context length: how many characters do we take to predict the next one?
X, Y = [], []
for w in words:
  #print(w)
  context = [0] * block_size
  for ch in w + '.':
    ix = stoi[ch]
    X.append(context)
    Y.append(ix)
    #print(''.join(itos[i] for i in context), '--->', itos[ix])
    context = context[1:] + [ix] # crop and append
X = torch.tensor(X)
Y = torch.tensor(Y)
print(f"Xshape:{X.shape}")
print(f"Yshape:{Y.shape}")
```

```python
# how to decide the starting learining rate ?
#--------------------------------------------------
# will do experiment with different lr rates:

# case1 lr very low 
## in that case loss is bearly decreasing

# Params (wight & baises) initalization:
#----------------------------------------------------------------
g = torch.Generator().manual_seed(2147483647) # for reproducibility
C = torch.randn((27, 2), generator=g)
W1 = torch.randn((6, 100), generator=g)
b1 = torch.randn(100, generator=g)
W2 = torch.randn((100, 27), generator=g)
b2 = torch.randn(27, generator=g)
parameters = [C, W1, b1, W2, b2]


paramss = sum(p.nelement() for p in parameters) # number of parameters in total
#print(f"{paramss=}")
for p in parameters:
  p.requires_grad = True

# Training loop:
#-----------------------------------------------
import time
start = time.time()
epochs = 50
batch_size = 32
losses = []
lr = 0.0001

for _ in range(epochs):

    
  ix = torch.randint(0, X.shape[0], (batch_size,))

  #forward pass:
  emb = C[X[ix]] # (32, 3, 2)
  h = torch.tanh(emb.view(-1, 6) @ W1 + b1) # (32, 100)
  logits = h @ W2 + b2 # (32, 27)
  loss = F.cross_entropy(logits, Y[ix])
  #print(f"loss: {loss}")
  losses.append(loss.item())
  # backwaed pass:
  for p in parameters:
    p.grad = None
  loss.backward()

  #wt updatae:
  for p in parameters:
    p.data += -lr * p.grad

end = round(time.time()-start, 2)

#---------------------------------
print(f"\n\nFinal results:\n")
print(f"-----------------------------------------")
print(f"total epochs: {epochs=}")
print(f"Total training time: {end} seconds")
print(f"final Loss: {loss.item()}")
print(f"Total Params:{paramss}")
print(f"-----------------------------------------")

# Plot loss vs epochs
plt.figure(figsize=(4, 3))  # Set the width and height in inches
plt.plot(losses)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```

```python
# evaluation on generation
# sample from the model
g = torch.Generator().manual_seed(2147483647 + 10)

for _ in range(20):
    
    out = []
    context = [0] * block_size # initialize with all ...
    while True:
      emb = C[torch.tensor([context])] # (1,block_size,d)
      h = torch.tanh(emb.view(1, -1) @ W1 + b1)
      logits = h @ W2 + b2
      probs = F.softmax(logits, dim=1)
      ix = torch.multinomial(probs, num_samples=1, generator=g).item()
      context = context[1:] + [ix]
      out.append(ix)
      if ix == 0:
        break
    
    print(''.join(itos[i] for i in out))
```
